```python
RAND = 123
PATH_TO_DATA = 'data/'
PATH_TO_SUBMIT = 'submissions/'
```

```python
GOOD = [
    'best_mean2',
    'submit_21',
    'lol2_2.txt',
    'SUPERmean',
    'SUPERmean',
    'SUPERmean',
    'lol2_2',
]
```

```python
pd_sum = pd.DataFrame(np.zeros(82797), columns=['target'], index=range(1, 82797 + 1))
count = 0
for sub in GOOD:
    if sub in os.listdir(PATH_TO_SUBMIT):
        pd_sum += pd.read_csv(PATH_TO_SUBMIT + sub, index_col='session_id')
        count += 1
mean = pd_sum / count
```
